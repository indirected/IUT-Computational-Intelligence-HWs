```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.array(
    [[1, 1, 0],
    [0, 0, 0],
    [1, 0, 1],
    [0, 1, 0]]
)
fig, ax = plt.subplots()
ax.scatter(a[a[:, 2] == 0, 0], a[a[:, 2] == 0, 1], s=300, label='0')
ax.scatter(a[a[:, 2] == 1, 0], a[a[:, 2] == 1, 1], s=300, label='1')
ax.axline([0.3, 0], slope=1, c='k')
ax.grid(visible=True, which='major', alpha=0.5)
ax.set_xlim((-0.25, 1.25))
ax.set_ylim((-0.25, 1.25))
ax.legend(prop={'size': 15})
fig.set_size_inches((12, 7))
fig.savefig('2_1.pdf', bbox_inches='tight')
```

```python
a = np.array(
    [[1, 2, 1],
    [2, 1, 1],
    [1, 1, 0],
    [1, 0, 0]]
)
fig, ax = plt.subplots()
ax.scatter(a[a[:, 2] == 0, 0], a[a[:, 2] == 0, 1], s=300, label='0')
ax.scatter(a[a[:, 2] == 1, 0], a[a[:, 2] == 1, 1], s=300, label='1')
ax.axline([1, 1.5], slope=-1, c='k')
ax.grid(visible=True, which='major', alpha=0.5)
ax.set_xlim((0.9, 2.15))
ax.set_ylim((-0.25, 2.25))
ax.legend(prop={'size': 15})
fig.set_size_inches((12, 7))
fig.savefig('3_1.pdf', bbox_inches='tight')
```
